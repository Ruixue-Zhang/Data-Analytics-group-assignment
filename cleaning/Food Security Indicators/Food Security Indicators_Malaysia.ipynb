```python
#read into dataset
import pandas as pd
df = pd.read_csv('Dataset Food Security Indicators_Malaysia.csv')
df.head()
```

```python
#find unique Item
uniques = df.Item.unique()[1:]
print(uniques)
```

```python
list_features = []
for item in df.Item.unique()[1:]:
    list_features.append(item)
```

```python
values_for_feature = {}
for l in list_features:
    vals = []
    filtered = df[df.Item == l]
    if list(filtered.Value) != []:
        for f in list(filtered.Value):
            vals.append(f)
        if len(vals) != 21:
            for i in range(21-len(vals)):
                vals.append('N/A')
            print(len(vals))
        values_for_feature[l] = vals  
```

```python
new_Malaysia = pd.DataFrame(data = values_for_feature)
```

```python
#put into dataframe
dataframes = [df]
print(len(dataframes))
```

```python
dataf=[]
```

```python
for i in range(len(dataframes)):
    values_for_feature = {}
    d = dataframes[i]
    for l in list_features:
        vals = []
        filtered = d[d.Item == l]
        for f in list(filtered.Value):
            vals.append(f)
        if len(vals) != 21:
            for j in range(21-len(vals)):
                vals.append(1)            
        values_for_feature[l] = vals
    g = pd.DataFrame(data = values_for_feature)
    dataf.append(g)
```

```python
#clean the data into Code Year columns, with specific attributes
dataf[0]
```

```python
full_dataframe = dataf[0]
for data in dataf[1:]:
    full_dataframe = full_dataframe.append(data)
```

```python
len(full_dataframe.index)
```

```python
#to identify any null values
full_dataframe.isnull().sum()
```

```python
#drop any na values
full_dataframe.dropna(inplace = True)
```

```python
#identify any which field has 1s
full_dataframe.head()
```

```python
#replacing the 1s with means of the columns:
import numpy as np
for c in full_dataframe.columns:
    col = full_dataframe[c]
    non_one = [float(i) for i in col if i != 1 and i[0] != '<']
    ave = np.mean(non_one)
    fixed = []
    for i in range(len(col)):
        if list(col)[i] == 1:
            fixed.append(ave)
        else:
            fixed.append(list(col)[i])        
    full_dataframe[c] = fixed
```
